```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 10 random samples
np.random.seed(42)
X = np.random.rand(10, 1)
y = 2 * X + 1 + np.random.randn(10, 1) * 0.1
```

```python
plt.scatter(X, y, color='b', label='Samples')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Initialize parameters
# theta = np.random.randn(2, 1)
theta = (2, 1)
X_b = np.c_[np.ones((10, 1)), X]
mu = 0.1  # Learning rate
n_iterations = 1000

# LMS Algorithm
for iteration in range(n_iterations):
    gradients = 2/10 * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - mu * gradients

# Print the final parameters
print("Final parameters:")
print(f"Intercept: {theta[0][0]:.4f}")
print(f"Slope: {theta[1][0]:.4f}")

# Plotting
plt.scatter(X, y, color='b', label='Samples')
plt.plot(X, X_b.dot(theta), color='r', label='LMS Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('LMS Learning with 10 Samples')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_samples = 10
num_iterations = 100
mu = 0.01  # Step size
lambda_ = 0.9  # Forgetting factor

# Generate synthetic data
np.random.seed(0)
x = np.random.randn(num_samples)  # Input signal
d = 2 * x + np.random.randn(num_samples) * 0.1  # Desired signal with noise

# Initialize weights and error array
w = np.zeros(num_samples)
error = np.zeros(num_iterations)

# CLMS algorithm
for n in range(num_iterations):
    for i in range(num_samples):
        # Calculate output
        y_hat = np.dot(w, x)
        
        # Calculate error
        e = d[i] - y_hat
        
        # Update weights with CLMS rule
        w += mu * e * x / (lambda_ + np.dot(x, x))
        
        # Store error for plotting
        error[n] += e**2

# Plotting results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(d, label='Desired Signal', marker='o')
plt.plot(np.dot(w, x), label='Output Signal', marker='x')
plt.title('Desired vs Output Signal')
plt.xlabel('Sample Index')
plt.ylabel('Signal Value')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(error, label='Error', color='red')
plt.title('Error over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error')
plt.legend()

plt.tight_layout()
plt.show()

print("Final Weights:", w)
```
